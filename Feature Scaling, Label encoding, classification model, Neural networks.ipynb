# Classification Model

```python
from sklearn.datasets import load_iris
iris = load_iris()

#feature matrix - x, response vector - y
x=iris.data
y=iris.target

#split x and y into training and test data sets 

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=1)
```

```python
#train the model
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train,y_train)

#making predictions
y_pred = gnb.predict(x_test)
```

```python
#compare- the actual respose values (y_test) and predicted response(y_pred)

from sklearn import metrics 
print("Gaussian NB model accuracy :", metrics.accuracy_score(y_test, y_pred)*100, "%")
```

```python
#activation function

def sigmoid(x):
    return (1/(1+np.exp(-x)))
```

```python
#creating feed forward NN
def f_forward(x, w1, w2):
    #hidden layer
    z1=x.dot(w1)
    a1=sigmoid(z1)
    
    #output layer
    z2=a1.dot(w2)
    a2=sigmoid(z2)
    
    return a2
```
